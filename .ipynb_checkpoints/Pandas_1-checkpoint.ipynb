# dropna( )

```python
#ignores the row with nan value
df.dropna()
#OR
#df.dropna(how = 'any')
```

```python
#ignores coloumn with nan value
df.dropna(axis=1)
#OR
#df.dropna(axis=1,how='any')
```

```python
df.dropna(how='all')
#if all the value in a row is nan then drops
```

```python
df.dropna(thresh = 2)
#has atleast 2 notnull value
```

```python
df.dropna(subset=['Product'])
#removes the whole row where nan is found in product column
```

```python
df.dropna(inplace=True)
df
#it also removes the row but doesn't create new data file it updates the old one 
```

# fillna( )

```python
df=pd.read_csv('C:\\Users\\Dell\\pandas\\sample_csv.csv',na_values={'Product':'Product2'})
df
```

```python
df.fillna(0)
#fills all nan value with 0
```

```python
df.fillna({'Product':'none','US Zip':0})
#fills all value with respective to dictionary
```

```python
df.fillna(method='ffill')
#df.fillna(method='pad') same output
#fills nan value with just before value in coloumn
```

```python
df.fillna(method='bfill',axis=1)
#fills with downvalue or after value axis 1 indicates horixontal value
```

```python
df.fillna(0,limit = 2)
#only 1st two element in each coloumn will change from nan to 0
```

```python
df.fillna(0, inplace=True)
df
#it doesnt create new dataframe rather modified the table
```

# replace( )

```python
df
```

```python
df.replace(to_replace = 0 , value={'Product':'Product2'})
#shortcut df.replace(0,{'product':'product2'})
#replace 0 with product2 of product coloumn using dict
```

```python
df.replace([1200,3600],0)
#more than one value into 1 value
```

```python
df.replace([1200,3600],[3600,1200])
#replace corredpondingly
```

```python
df.replace('[A-Za-z]',0,regex=True)
#regex is used to change string value in integer and '[A-Za-z]' means all string value
```

```python
df
```

```python
df.replace('Visa',method='ffill')
#visa replaced by forward or before value
```

```python
df.replace('Visa',method='ffill',limit=1)
#limits first 1 value
```

```python
df.replace({'Product':0},'Product2',inplace=True)
df
#inplace modifies the dataframe rather than creating new
```
